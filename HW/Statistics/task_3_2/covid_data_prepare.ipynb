Перед вами датасет по заболеваниям Covid-19. Мы предобработали данные, чтобы вам было проще с ними работать. 


```python
from covid import Covid
import pandas as pd
import datetime
import seaborn as sns

#we used John Hopkins dataset
#But it also contains non-countries values
covid_hopkins = Covid(source="john_hopkins")
covid_world = Covid(source="worldometers")
```

```python
data_hopkins = pd.DataFrame(covid_hopkins.get_data())
data_hopkins.last_update = data_hopkins.last_update / 1e3
data_hopkins = data_hopkins.drop(["active", "recovered"], axis=1)
data_hopkins = data_hopkins.astype({"id" : "int32", "country" : "str", 
                     "confirmed" : "int64", "deaths" : "int64", "latitude" : "float", 
                    "longitude" : "float", "last_update" : "int64"})
data_hopkins.last_update = pd.to_datetime(data_hopkins.last_update, unit="s")
data_hopkins = data_hopkins.sort_values(by=["confirmed"], ascending=False)
data_hopkins
```
